```python
import pandas as pd
import numpy as np
import geopandas as gpd
```

```python
dfAll = pd.read_excel("C:\\Users\\prave\\Repos\\ACModelling\\Data\\ApartmentRentals\\ApTRentAll.xlsx")
dfAll2 = pd.read_excel("C:\\Users\\prave\Repos\\ACModelling\\Data\\ApartmentRentals\\AptRentAll2.xlsx")
dfAC = pd.read_excel("C:\\Users\\prave\\Repos\\ACModelling\\Data\\ApartmentRentals\\AptRentAC.xlsx")
```

```python
dfA = pd.concat([dfAll, dfAll2], ignore_index=True)
dfA = dfA.drop_duplicates()
dfA = dfA.drop_duplicates(subset=['Property URL'])
dfA
```

```python
dfAC = dfAC.drop_duplicates()
dfAC
```

```python
# Getting unique 'Property URL' values from dfAC and dfA
unique_urls_dfAC = set(dfAC['Property URL'].unique())
unique_urls_dfA = set(dfA['Property URL'].unique())
```

```python
urls_not_in_dfA = unique_urls_dfAC.difference(unique_urls_dfA)

# Counting the number of unique URLs in dfAC that are not in dfA
count_not_in_dfA = len(urls_not_in_dfA)

# Printing the count
print(f"Number of unique 'Property URLs' in dfAC that are not in dfA: {count_not_in_dfA}")
```

```python
urls_not_in_dfAC = unique_urls_dfA.difference(unique_urls_dfAC)

count_not_in_dfAC = len(urls_not_in_dfAC)

print(f"Number of unique 'Property URLs' in dfA that are not in dfAC: {count_not_in_dfAC}")
```

```python
dfAC = dfAC[['Property URL']].drop_duplicates()  # Ensuring uniqueness
dfAC ['AC'] = 1

dfA = dfA.merge(dfAC, on='Property URL', how='left')

dfA['AC'].fillna(0, inplace=True)
```

```python
ac_counts = dfA['AC'].value_counts()
print(ac_counts)
```
