```python
import pandas as pd
import geopandas as gpd
from shapely import wkt
from shapely.geometry import Point, Polygon
```

```python
geocode_results = pd.read_csv('C:\\Users\\prave\\Repos\\ACModelling\\Data\\ApartmentRentals\\geocoderesult.csv', header=None, index_col=False, names= ['Unique ID', 'Address', 'Match','Exact','Formatted Address','Coordinates','N','LR'] )
geocode_results = geocode_results[pd.to_numeric(geocode_results['Unique ID'], errors='coerce').notnull()]
geocode_results['Unique ID'] = geocode_results['Unique ID'].astype('int64')
```

```python
geocode_results
```

```python
dfA = pd.read_csv("C:\\Users\\prave\\Repos\\ACModelling\\Data\\ApartmentRentals\\AptRentClean.csv")
dfA = pd.merge(dfA, geocode_results, on='Unique ID', how='left')
```

```python
dfA.columns
```

```python
dfA = dfA[['Unique ID', 'Property URL', 'Property type', 'Rental price (USD per month)','Living area','Living area unit',
       'Price per living area unit (USD)', 'Bedrooms', 'Bathrooms', 'Formatted Address', 'Coordinates' ]]
```

```python
dfA
```

```python
dfA['wkt'] = dfA['Coordinates'].apply(lambda x: 'POINT (' + x.split(',')[0] + ' ' + x.split(',')[1] + ')' if isinstance(x, str) else None)
```

```python
dfA['geometry'] = dfA['wkt'].apply(lambda wkt_value: wkt.loads(wkt_value) if wkt_value is not None else None)
```

```python
dfA = dfA.drop(['Coordinates', 'wkt'], axis=1)
```

```python
gdf = gpd.GeoDataFrame(dfA, geometry='geometry')
```

```python
gdf = gdf.set_crs('EPSG:4326')
```

```python
gdf_explore = gdf.dropna(subset=['geometry'])
gdf_explore.explore()
```
